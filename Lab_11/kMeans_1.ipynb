```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
# print(data)
```

```python
data = pd.read_csv('../datasets/Mall_Customers.csv')
```

```python
from sklearn.cluster import KMeans

x = data.drop(columns=['CustomerID','Genre','Age'], axis=1).values

plt.figure(figsize=(15,8))
plt.scatter(x[:,0],x[:,1])
plt.xlabel('ANNUAL INCOME')
plt.ylabel('SPENDING SCORE')
plt.show()

wcss=[]
for i in range(1,15):
    kmeans = KMeans(n_clusters=i,init='k-means++', random_state=2)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
    
plt.figure(figsize=(15,8))
plt.plot(range(1,15), wcss)
plt.title('ELBOW POINT GRAPH')
plt.xlabel('NO. OF CLUSTER(K)')
plt.ylabel('WCSS')
plt.show()
```

```python

y=kmeans.fit_predict(x)

data['KMean'] = kmeans.labels_
data
print(kmeans.cluster_centers_)


plt.figure(figsize=(15, 8))
plt.subplot(1, 2, 1) 
plt.scatter(x[:, 0], x[:, 1])
plt.xlabel('ANNUAL INCOME')
plt.ylabel('SPENDING SCORE')
plt.title('Scatter Plot')
```

```python
plt.subplot(1, 2, 2)  
plt.scatter(x[y == 0, 0], x[y == 0, 1], s=50, c='red', label='CLUSTER 1')
plt.scatter(x[y == 1, 0], x[y == 1, 1], s=50, c='yellow', label='CLUSTER 2')
plt.scatter(x[y == 2, 0], x[y == 2, 1], s=50, c='blue', label='CLUSTER 3')
plt.scatter(x[y == 3, 0], x[y == 3, 1], s=50, c='orange', label='CLUSTER 4')
plt.scatter(x[y == 4, 0], x[y == 4, 1], s=50, c='green', label='CLUSTER 5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='black', label='CENTROID')
plt.title('Customer Cluster Using KMeans')
plt.xlabel('ANNUAL INCOME')
plt.ylabel('SPENDING SCORE')


plt.tight_layout()


plt.show()


plt.figure(figsize=(20, 4))

for cluster_id in range(5):
    plt.subplot(1, 5, cluster_id + 1)
    plt.scatter(x[y == cluster_id, 0], x[y == cluster_id, 1], s=50, label=f'CLUSTER {cluster_id + 1}')
    plt.scatter(kmeans.cluster_centers_[cluster_id, 0], kmeans.cluster_centers_[cluster_id, 1], s=200, c='white', label='CENTROID')
    plt.title(f'Cluster {cluster_id + 1}')
    plt.xlabel('ANNUAL INCOME')
    plt.ylabel('SPENDING SCORE')


plt.tight_layout()


plt.show()
```
