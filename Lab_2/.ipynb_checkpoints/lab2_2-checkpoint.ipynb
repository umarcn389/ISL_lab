```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math as m
```

```python
x=[1,2,3,4,5]
y=[10,11,12,13,14]
a=[5,6,7,8]
x1=np.array(a)
y1=x1**2
plt.scatter(x,y)
plt.plot(x,y,"g--",label="line")
plt.scatter(x1,y1)
plt.plot(x1,y1,"b--",label="x^2")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```
